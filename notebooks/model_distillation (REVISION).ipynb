```python
from xgboost2.clinical_validation import read_data
```

```python
# model 1
model = 1
model_version_mort = 22

data_dict = read_data(model = model, 
                      out = 3, # mortality
                      version = model_version_mort)

labels = ['BGbinary', 'URESE', 'ALBSE', 'CKDESE', 'BGA',
          'TRC', 'Geboortedatum', 'TEISE', 'AFSE', 'GLUVB',
          'GLUAB', 'LDSE', 'LACAB', 'KRESE', 'LPSSE',
          'TBISE', 'GGTSE', 'CRPSE', 'MAGSE', 'HB']
print(len(labels))
```

```python
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
import pandas as pd 
import numpy as np
from sklearn.metrics import roc_curve, auc
from math import sqrt

top_n = [20, 15, 10, 5, 3, 1]
cv = StratifiedKFold(n_splits=5)
aucs = {}

x = pd.DataFrame(data_dict['x'], columns=data_dict['features'])
y = data_dict['y']

for n in top_n:
    _aucs = []
    model = XGBClassifier()
    _x = x[labels[0:n]].values
    
    for train, test in cv.split(_x, y):
        model.fit(_x[train], y[train])
        
        probas_ = model.predict_proba(_x[test])
        fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
        _aucs.append(auc(fpr, tpr))
        
    mean_auc = np.mean(_aucs)
    _ci = 1.96 * (np.std(_aucs) / sqrt(5))
    aucs[str(n)] = "{:.3f} ({:.3f}-{:.3f})".format(mean_auc,
                                              mean_auc - _ci,
                                              mean_auc + _ci)
    print(aucs)
```
